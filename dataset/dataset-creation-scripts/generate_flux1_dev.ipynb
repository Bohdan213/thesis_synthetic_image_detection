```python
pipe.set_progress_bar_config(disable=True)

from tqdm import tqdm

for idx in range(10):
  for image_class in tqdm(topics):

      prompt = image_class["topic"] + ", photography, professional"

      image = pipe(
          prompt,
          negative_prompt="painting, drawing, illustration, abstract",
          num_images_per_prompt=1,
          height=1024,
          width=1024,
          guidance_scale=4,
          num_inference_steps=50,
          max_sequence_length=512
      ).images[0]

      image.save(f'FLUX1-dev/FLUX1-dev_class_{image_class["id"]}_id_{idx}.png')
```
